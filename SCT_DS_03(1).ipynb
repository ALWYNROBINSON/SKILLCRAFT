```python
from sklearn.model_selection import train_test_split
print(4119*0.25)
```

```python
xt,xtst,yt,ytst=train_test_split(a,b,test_size=0.25,random_state=1)
print(xt.shape)
print(xtst.shape)
print(yt.shape)
print(ytst.shape)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

def em(ytst,yp):
    acc=accuracy_score(ytst,yp)
    print("ACCURACY SCORE:",acc)
    cm=confusion_matrix(ytst,yp)
    print("CONFUSION MATRIX:\n",cm)
    print('CLASSIFICATION REPORT\n',classification_report(ytst,yp))

def mscore(model):
    ts=model.score(xt,yt)
    tsts=model.score(xtst,ytst)
    print('TRAINING SCORE:',ts)
    print('TESTING SCORE:',tsts)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dtc.fit(xt,yt)
```

```python
mscore(dtc)
```

```python
ypdtc=dtc.predict(xtst)
ypdtc
```

```python
em(ytst,ypdtc)
```
