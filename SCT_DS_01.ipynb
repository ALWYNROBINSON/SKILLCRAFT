```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n_samples = 1000
ages = np.random.normal(loc=35, scale=10, size=n_samples)  # Ages centered around 35 with std dev 10
ages = np.clip(ages, 0, 100).astype(int)  # Ensure ages are between 0 and 100
genders = np.random.choice(['Male', 'Female', 'Other'], size=n_samples, p=[0.48, 0.48, 0.04])
data = pd.DataFrame({'Age': ages, 'Gender': genders})
```

```python
plt.figure(figsize=(10, 5))
plt.hist(data['Age'], bins=20, color='#1f77b4', edgecolor='black')
plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(8, 5))
gender_counts.plot(kind='bar', color=['#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Gender Distribution in Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()
```
